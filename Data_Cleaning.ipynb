```
import io
import os
import re
import zipfile
import numpy as np
import pandas as pd
from google.colab import files

# ============================================================
# 1) CONFIGURATION
# ============================================================

CONFIG = {
    "Inward": {
        "detect_keywords": ["inward"],
        "detect_columns": ["Inward Code", "Received material from", "Project name"],
        "keep_columns": [
            "Inward Code",
            "Date",
            "Received Material From",
            "Vendor Location",
            "Source",
            "Project Name",
            "Vehicle Number",
            "Loading Cost",
            "Driver Name",
            "Vehicle Vendor Name",
            "Transportation Cost",
            "Additional Cost",
            "Total Tax",
            "Material",
            "Received Quantity",
            "Rate",
            "Value of Received Material",
            "Rejected Quantity",
            "Accepted Quantity",
            "Value of Accepted Material",
            "Net Procurement Cost",
            "Payment Status",
            "Weighment Slip",
            "Vendor Invoice",
            "Vehicle GPS Photo",
            "Inward Record Photo",
            "Material GPS Photo",
        ],
        "non_ffill_columns": {
            "Material",
            "Received Quantity",
            "Rate",
            "Amount",
            "Rejected Quantity",
            "Rejected Quantity %",
            "Accepted Billed Qty",
            "Value of Accepted Qty",
        },
        "aliases": {
            "inward code": "Inward Code",
            "date": "Date",
            "received material from": "Received Material From",
            "vendor location": "Vendor Location",
            "source": "Source",
            "project name": "Project Name",
            "vehicle number": "Vehicle Number",
            "loading cost": "Loading Cost",
            "driver name": "Driver Name",
            "vehicle vendor name": "Vehicle Vendor Name",
            "transportation cost": "Transportation Cost",
            "transportation cost in rs": "Transportation Cost",
            "additional cost": "Additional Cost",
            "additional cost in rs": "Additional Cost",
            "total tax": "Total Tax",
            "total tax in": "Total Tax",
            "material": "Material",
            "received quantity": "Received Quantity",
            "received quantity in kg": "Received Quantity",
            "rate": "Rate",
            "rate in rs": "Rate",
            "amount": "Amount",
            "amount in rs": "Amount",
            "rejected qty": "Rejected Quantity",
            "rejected qty in kg": "Rejected Quantity",
            "rejected quantity": "Rejected Quantity",
            "payment status": "Payment Status",
            "weighment slip": "Weighment Slip",
            "vendor invoice": "Vendor Invoice",
            "vehicle gps photo": "Vehicle GPS Photo",
            "inward record photo": "Inward Record Photo",
            "inward record": "Inward Record Photo",
            "material gps photo": "Material GPS Photo",
            "material gps": "Material GPS Photo",

            # extra original columns
            "rejected qty in": "Rejected Quantity %",
            "accepted billed qty in kg": "Accepted Billed Qty",
            "value of accepted qty in rs": "Value of Accepted Qty",
            "total received qty in kg": "Total Received Qty",
            "total accepted qty in kg": "Total Accepted Qty",
            "total rejected qty in kg": "Total Rejected Qty",
            "total amount in rs": "Total Amount",
            "dc number": "DC Number",
            "invoice number": "Invoice Number",
            "s no": "S.no",
            "sno": "S.no",
            "s.no": "S.no",
        },
    },

    "Production": {
        "detect_keywords": ["production"],
        "detect_columns": ["Production Code", "Process-Equipment", "No. of Present Staffs"],
        "keep_columns": [
            "Production Code",
            "Date",
            "Shift",
            "Process-Equipment",
            "Time Taken in Hrs",
            "No. of Staff Present",
            "Material",
            "Quantity in Kg",
            "Total Rejection in Kg",
            "Materials",
            "Material From",
            "Material Quantity",
        ],
        "non_ffill_columns": {
            "Material",
            "Quantity in Kg",
            "Total Rejection in Kg",
            "Materials",
            "Material From",
            "Material Quantity",
        },
        "aliases": {
            "production code": "Production Code",
            "date": "Date",
            "shift": "Shift",
            "process equipment": "Process-Equipment",
            "process-equipment": "Process-Equipment",
            "time taken in hrs": "Time Taken in Hrs",
            "time taken in hrs.": "Time Taken in Hrs",
            "time taken in": "Time Taken in Hrs",
            "time taken in hrs ": "Time Taken in Hrs",
            "time taken in hrs)": "Time Taken in Hrs",
            "time taken (in hrs)": "Time Taken in Hrs",
            "no of present staffs": "No. of Staff Present",
            "no. of present staffs": "No. of Staff Present",
            "no. of staff present": "No. of Staff Present",
            "material": "Material",
            "quantity in kg": "Quantity in Kg",
            "quantity (in kg)": "Quantity in Kg",
            "total rejection in kg": "Total Rejection in Kg",
            "total rejection (in kg)": "Total Rejection in Kg",
            "materials": "Materials",
            "material from": "Material From",
            "material quantity": "Material Quantity",
            "material quantity in kg": "Material Quantity",
            "material quantity (in kg)": "Material Quantity",

            # extra original columns
            "total quantity in kg": "Total Quantity",
            "total quantity (in kg)": "Total Quantity",
            "description": "Description",
            "s no": "S.no",
            "sno": "S.no",
            "s.no": "S.no",
        },
    },

    "Outward": {
        "detect_keywords": ["outward"],
        "detect_columns": ["Outward Code", "Vehicle number", "Dispatched Qty (in KG)"],
        "keep_columns": [
            "Outward Code",
            "Date",
            "Project",
            "Customer",
            "Destination",
            "Vehicle Number",
            "Transport Vendor",
            "Driver Name",
            "Driver Contact No",
            "Loading Cost",
            "Additional Transport Cost",
            "Transportation Cost",
            "Material",
            "Dispatched Quantity",
            "Rate",
            "Value of Dispatched Material",
            "Rejected Quantity",
            "Accepted Quantity",
            "Value of Accepted Material",
            "Tax",
            "Payment Status",
            "Incentive Cost per Kg",
            "Total Incentive Cost",
            "Net Material Sales Cost",
            "GPS Photo 1",
            "GPS Photo 2",
            "Invoice",
            "LR Copy",
            "DC Challan",
            "Weighment Slip",
            "E-Way Bill",
            "Received Acknowledgement Form ED",
            "Vehicle and Material Photo",
        ],
        "non_ffill_columns": {
            "Material",
            "Dispatched Quantity",
            "Rate",
            "Billed Amount",
            "Rejected Quantity",
            "Tax",
            "Incentive Cost per Kg",
        },
        "aliases": {
            "outward code": "Outward Code",
            "date": "Date",
            "project": "Project",
            "customer": "Customer",
            "destination": "Destination",
            "vehicle number": "Vehicle Number",
            "vehicle no": "Vehicle Number",
            "transport vendor": "Transport Vendor",
            "driver name": "Driver Name",
            "driver contact no": "Driver Contact No",
            "loading cost": "Loading Cost",
            "additional transport cost": "Additional Transport Cost",
            "transportation cost": "Transportation Cost",
            "transport cost": "Transportation Cost",
            "transport cost in rs": "Transportation Cost",
            "material": "Material",
            "dispatched quantity": "Dispatched Quantity",
            "dispatched qty in kg": "Dispatched Quantity",
            "dispatched qty": "Dispatched Quantity",
            "rate": "Rate",
            "rate in rs": "Rate",
            "billed amount": "Billed Amount",
            "billed amount in rs": "Billed Amount",
            "rejected quantity": "Rejected Quantity",
            "rejection qty in kg": "Rejected Quantity",
            "rejection qty": "Rejected Quantity",
            "tax": "Tax",
            "total tax in": "Tax",
            "payment status": "Payment Status",
            "incentive cost per kg": "Incentive Cost per Kg",
            "incentive costkg": "Incentive Cost per Kg",
            "gps photo 1": "GPS Photo 1",
            "gsp photo 1": "GPS Photo 1",
            "gps photo 2": "GPS Photo 2",
            "gsp photo 2": "GPS Photo 2",
            "invoice": "Invoice",
            "lr copy": "LR Copy",
            "dc challan": "DC Challan",
            "weighment slip": "Weighment Slip",
            "e-way bill": "E-Way Bill",
            "eway bill": "E-Way Bill",
            "received acknowledgement form ed": "Received Acknowledgement Form ED",
            "received acknowledgement from ed": "Received Acknowledgement Form ED",
            "vehicle and material photo": "Vehicle and Material Photo",
            "vehicle and material": "Vehicle and Material Photo",

            # extra original columns
            "address line 1": "Address Line 1",
            "address line 2": "Address Line 2",
            "pincode": "Pincode",
            "city": "City",
            "state": "State",
            "country": "Country",
            "type": "Type",
            "total accepted qty in kg": "Total Accepted Qty",
            "value of accepted material in rs": "Value of Accepted Material (Old)",
            "rejection %": "Rejection %",
            "total dispatched qty in kg": "Total Dispatched Qty",
            "total value of dispatched materials in rs": "Total Value of Dispatched Materials",
            "total value of accepted qty in rs": "Total Value of Accepted Qty",
            "total incentive cost in rs": "Total Incentive Cost (Old)",
            "description": "Description",
            "wma": "WMA",
            "dc number": "DC Number",
            "invoice number": "Invoice Number",
            "s no": "S.no",
            "sno": "S.no",
            "s.no": "S.no",
        },
    },
}


# ============================================================
# 2) HELPER FUNCTIONS
# ============================================================

def canonical_col(name: str) -> str:
    """
    Normalise a column name so that similar spellings map together.
    """
    if name is None:
        return ""
    s = str(name).strip().lower()
    s = s.replace("_", " ").replace("-", " ")
    s = s.replace("/", " ")
    s = s.replace(".", " ")
    s = s.replace("(", " ").replace(")", " ")
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s


def build_alias_lookup(dataset_name: str):
    aliases = CONFIG[dataset_name]["aliases"]
    lookup = {}
    for raw_key, target in aliases.items():
        lookup[canonical_col(raw_key)] = target
    return lookup


def detect_dataset(filename: str, columns):
    fname = filename.lower()

    for dataset_name, cfg in CONFIG.items():
        if any(k in fname for k in cfg["detect_keywords"]):
            return dataset_name

    colset = set(columns)
    for dataset_name, cfg in CONFIG.items():
        hits = sum(1 for c in cfg["detect_columns"] if c in colset)
        if hits >= 1:
            return dataset_name

    canon_cols = set(canonical_col(c) for c in columns)
    for dataset_name, cfg in CONFIG.items():
        detect_canon = set(canonical_col(c) for c in cfg["detect_columns"])
        hits = len(canon_cols.intersection(detect_canon))
        if hits >= 1:
            return dataset_name

    return None


def standardise_columns(df: pd.DataFrame, dataset_name: str) -> pd.DataFrame:
    lookup = build_alias_lookup(dataset_name)

    new_cols = []
    for col in df.columns:
        canon = canonical_col(col)
        std = lookup.get(canon, col)
        new_cols.append(std)

    df = df.copy()
    df.columns = new_cols
    return df


def coalesce_duplicate_columns(df: pd.DataFrame) -> pd.DataFrame:
    if df.columns.is_unique:
        return df

    out = pd.DataFrame(index=df.index)
    for col in df.columns.unique():
        same = df.loc[:, df.columns == col]

        if same.shape[1] == 1:
            out[col] = same.iloc[:, 0]
        else:
            combined = same.iloc[:, 0].copy()
            for j in range(1, same.shape[1]):
                combined = combined.combine_first(same.iloc[:, j])
            out[col] = combined

    return out


def apply_forward_fill(df: pd.DataFrame, dataset_name: str) -> pd.DataFrame:
    df = df.copy()
    non_ffill = set(CONFIG[dataset_name]["non_ffill_columns"])
    fill_cols = [c for c in df.columns if c not in non_ffill]

    if fill_cols:
        df[fill_cols] = df[fill_cols].ffill()

    return df


def to_numeric_safe(series):
    return pd.to_numeric(series, errors="coerce").fillna(0)


def split_cost_equally_by_code(df: pd.DataFrame, code_col: str, cost_cols: list) -> pd.DataFrame:
    """
    Split the given cost columns equally across all rows belonging to the same code.
    This OVERWRITES the original cost columns with split values.
    """
    df = df.copy()

    if code_col not in df.columns:
        return df

    # Make sure code column is usable
    code_series = df[code_col].astype(str).str.strip()

    # Count rows per code
    row_count = code_series.groupby(code_series).transform("size").replace(0, 1)

    for col in cost_cols:
        if col not in df.columns:
            df[col] = 0

        numeric_col = to_numeric_safe(df[col])

        # Take one original code-level cost (first non-null / first row after ffill)
        code_level_cost = numeric_col.groupby(code_series).transform("first")

        # Overwrite same column with equally split cost
        df[col] = (code_level_cost / row_count).round(2)

    return df


def add_derived_columns(df: pd.DataFrame, dataset_name: str) -> pd.DataFrame:
    """
    Add calculated business columns for Inward and Outward.
    Shared costs are split equally across rows of the same code.
    """
    df = df.copy()

    if dataset_name == "Inward":
        for col in [
            "Received Quantity",
            "Rate",
            "Rejected Quantity",
            "Transportation Cost",
            "Additional Cost",
            "Loading Cost",
            "Inward Code",
        ]:
            if col not in df.columns:
                df[col] = 0

        # Split cost columns equally by Inward Code
        df = split_cost_equally_by_code(
            df,
            code_col="Inward Code",
            cost_cols=["Transportation Cost", "Additional Cost", "Loading Cost"]
        )

        received_qty = to_numeric_safe(df["Received Quantity"])
        rate = to_numeric_safe(df["Rate"])
        rejected_qty = to_numeric_safe(df["Rejected Quantity"])
        transportation_cost = to_numeric_safe(df["Transportation Cost"])
        additional_cost = to_numeric_safe(df["Additional Cost"])
        loading_cost = to_numeric_safe(df["Loading Cost"])

        df["Accepted Quantity"] = (received_qty - rejected_qty).clip(lower=0)
        df["Value of Received Material"] = (received_qty * rate).round(2)
        df["Value of Accepted Material"] = (df["Accepted Quantity"] * rate).round(2)

        df["Net Procurement Cost"] = (
            df["Value of Accepted Material"]
            - transportation_cost
            - additional_cost
            - loading_cost
        ).round(2)

    elif dataset_name == "Outward":
        for col in [
            "Dispatched Quantity",
            "Rate",
            "Rejected Quantity",
            "Transportation Cost",
            "Additional Transport Cost",
            "Loading Cost",
            "Outward Code",
            "Incentive Cost per Kg",
        ]:
            if col not in df.columns:
                df[col] = 0

        # Split cost columns equally by Outward Code
        df = split_cost_equally_by_code(
            df,
            code_col="Outward Code",
            cost_cols=["Transportation Cost", "Additional Transport Cost", "Loading Cost"]
        )

        dispatched_qty = to_numeric_safe(df["Dispatched Quantity"])
        rate = to_numeric_safe(df["Rate"])
        rejected_qty = to_numeric_safe(df["Rejected Quantity"])
        transportation_cost = to_numeric_safe(df["Transportation Cost"])
        additional_transport_cost = to_numeric_safe(df["Additional Transport Cost"])
        loading_cost = to_numeric_safe(df["Loading Cost"])
        incentive_per_kg = to_numeric_safe(df["Incentive Cost per Kg"])

        df["Accepted Quantity"] = (dispatched_qty - rejected_qty).clip(lower=0)
        df["Value of Dispatched Material"] = (dispatched_qty * rate).round(2)
        df["Value of Accepted Material"] = (df["Accepted Quantity"] * rate).round(2)
        df["Total Incentive Cost"] = (incentive_per_kg * df["Accepted Quantity"]).round(2)

        df["Net Material Sales Cost"] = (
            df["Value of Accepted Material"]
            + df["Total Incentive Cost"]
            - transportation_cost
            - additional_transport_cost
            - loading_cost
        ).round(2)

    return df


def keep_only_required_columns(df: pd.DataFrame, dataset_name: str) -> pd.DataFrame:
    df = df.copy()
    keep_cols = CONFIG[dataset_name]["keep_columns"]

    for col in keep_cols:
        if col not in df.columns:
            df[col] = pd.NA

    df = df[keep_cols]
    return df


def clean_dataset(df: pd.DataFrame, dataset_name: str) -> pd.DataFrame:
    """
    Full cleaning pipeline for one dataset.
    """
    df = standardise_columns(df, dataset_name)
    df = coalesce_duplicate_columns(df)
    df = apply_forward_fill(df, dataset_name)
    df = add_derived_columns(df, dataset_name)
    df = keep_only_required_columns(df, dataset_name)
    df = df.loc[:, ~df.columns.duplicated()]
    return df


# ============================================================
# 3) UPLOAD + PROCESS
# ============================================================

print("📂 Upload one or more CSV files (Inward / Production / Outward)...")
uploaded = files.upload()

cleaned_files = []

for filename, content in uploaded.items():
    try:
        print(f"\n🔍 Processing: {filename}")

        df = pd.read_csv(io.BytesIO(content))

        dataset_name = detect_dataset(filename, df.columns)
        if not dataset_name:
            print(f"⚠️ Could not detect dataset type for: {filename}")
            print("   Skipping this file.")
            continue

        print(f"✅ Detected dataset: {dataset_name}")
        print(f"   Original shape: {df.shape[0]} rows × {df.shape[1]} columns")

        cleaned_df = clean_dataset(df, dataset_name)

        print(f"   Cleaned shape:  {cleaned_df.shape[0]} rows × {cleaned_df.shape[1]} columns")
        print(f"   Final columns: {list(cleaned_df.columns)}")

        print("\n🧪 Sample cleaned output:")
        display(cleaned_df.head(10))

        output_filename = os.path.splitext(filename)[0] + "_cleaned.csv"
        cleaned_df.to_csv(output_filename, index=False)
        cleaned_files.append(output_filename)

        print(f"💾 Saved: {output_filename}")

    except Exception as e:
        print(f"❌ Error processing {filename}: {e}")


# ============================================================
# 4) ZIP ALL CLEANED FILES + DOWNLOAD
# ============================================================

if cleaned_files:
    zip_name = "cleaned_datasets.zip"
    with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as zf:
        for f in cleaned_files:
            zf.write(f)

    print(f"\n📦 Created ZIP: {zip_name}")
    files.download(zip_name)

else:
    print("\n⚠️ No cleaned files were generated.")
```
